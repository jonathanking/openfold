```python
# Load csv in pandas
test_csv = "out/evaluation/230420/xfer_scnmin_omm_04_eval_s1659/test_results.csv"
# test_csv = "out/evaluation/230420/xfer_scnmin_omm_04_eval_s1659/val_results.csv"
df = pd.read_csv(test_csv)
df.sort_values(by='gdcall_aa', ascending=False).head()
```

```python
# Make a column that multiplies the negative of openmm_raw_energy by 1-gdcall_aa
# The higher the value of this column the more likely we have a good energy but low accuracy
df['goodenergy_lowacc'] = -df['openmm_raw_energy'] * (1 - df['gdcall_aa'])
```

```python
# Sort by goodenergy_lowacc, descending
df.sort_values(by='goodenergy_lowacc', ascending=False).head(10)
```

```python
# For the top 10, copy the prediction to jk_research_evaluations/230420/investigate_pdbs
# We need to copy both the prediction (out/evaluation/230420/xfer_scnmin_omm_04_eval_s1659/pdbs/pred/pred*{protein_name}.pdb)
# and the true (out/evaluation/230420/xfer_scnmin_omm_04_eval_s1659/pdbs/true/true*{protein_name}.pdb)
# to the investigate_pdbs folder

# Copy the predictions
for protein_name in df.sort_values(by='goodenergy_lowacc', ascending=False).head(10)['protein_name']:
    !cp out/evaluation/230420/xfer_scnmin_omm_04_eval_s1659/pdbs/pred/pred*{protein_name}.pdb jk_research/evaluations/230420/investigate_pdbs/
    !cp out/evaluation/230420/xfer_scnmin_omm_04_eval_s1659/pdbs/true/true*{protein_name}.pdb jk_research/evaluations/230420/investigate_pdbs/
```
